```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
print(x)
y=x**2
print(y)
```

```python
plt.figure(figsize=[5,10])
#plt.figure()
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(-5,5,11)
y=x**2
# print(x)
# print(y)
plt.figure(figsize=[3,5],facecolor='r')
plt.plot(x,y,color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('square of x')
plt.grid(True)
plt.axis([-5,5,0,25])
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.figure(figsize=[8,6], facecolor = 'r')
plt.plot(x,y, 'k',alpha = 0.8 , linestyle = '-.',linewidth=4, marker="o")
```

```python
plt.figure()
plt.plot(x,y,'o')
plt.show()
```

- subpot

```python
x=np.linspace(0,5,7)
y=x**2
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g-')
```

```python
x=np.linspace(0,5,7)
y=x**2
plt.figure()
plt.subplot(2,3,1)
plt.plot(x,y,'r--')

plt.subplot(2,3,2)
plt.plot(y,x,'y')

plt.subplot(2,3,3)
plt.plot(y,x,'c')

plt.subplot(2,3,4)
plt.plot(y,x,'g*')

plt.subplot(2,3,5)
plt.plot(y,x,'b')

plt.subplot(2,3,6)
plt.plot(y,x,'k')
```

```python
import numpy as np
x=np.linspace(0,5,11)


y=x**2
y1=x1**3
plt.figure(figsize=[10,10],facecolor='y')
plt.plot(x,y,'r',label='power of 2')
plt.plot(x,y1,'g',label='power of 3')
plt.legend(loc=1)
plt.grid(color='k',alpha=0.5)
```

```python
x=[5,10,8]
y=[12,6,15]

x1=[4,12,8]
y1=[10,2,14]
plt.scatter(x,y,color='r',label='Red',marker='*',s=72,alpha=0.9)
plt.scatter(x1,y1,color='g',label='green')
plt.legend(loc=3)

plt.grid()
```

```python
x=[5,10,8]
y=[12,6,15]

x1=[4,12,8]
y1=[10,2,14]

plt.barh(x,y,color='r',label='red',height=1.9,align='center')
plt.bar(x1,y1,color='g',label='green',width=0.5,align='edge')
plt.legend()
plt.grid()
```

```python
x=[5,10,8]
y=[12,6,15]

x1=[4,12,8]
y1=[10,2,14]

plt.bar(x,y,color='r',label='red',width=1,align='center')
plt.bar(x1,y1,color='g',label='green',width=0.5,align='edge')
plt.legend()
plt.grid()
```
