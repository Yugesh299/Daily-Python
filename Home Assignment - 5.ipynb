```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
df=pd.read_csv("ign.csv")
df1=df.drop(['url','Unnamed: 0','score_phrase'],axis=1,inplace = True)
#df
df.head()
```

```python
#Q1
number=df['score']>7
q=dict(df[number].count())
print("Number of games having more than 7 score are:{0}".format(q['score']))
```

```python
#Q2
j=list(df['editors_choice'].value_counts())
print("Number of games which got the editor's choice award:{0}".format(j[1]))
df[df['editors_choice']=='Y'];
```

```python
#Q3
maxscore=df[['genre','score']]
maximum=df['score']
for name,value in maxscore.groupby('genre'):
    print(name,maximum.max())
```

```python
#Q4
for name,values in df.groupby('title'):
    h=df['score'].max()
    if h==10:
        print(name)
    else:
        break
```

```python
#Q5
platform_maximum=df[['platform','score']]
for names,values in platform_maximum.groupby('platform'):
    print(names)
```

```python
#Q6
column_xbox = df["platform"] == 'Xbox'
final_xbox=df[column_xbox][['platform','score']]
print("The Average score of Xbox platform is {0}".format(final_xbox['score'].mean()))
```

```python
#Q7
df['release_year'].unique()
max_count=df['release_year'].value_counts().max()
mrelease_year=dict(df['release_year'].value_counts())
print("The most releases happend in the year {0} with a total of {1} games released".format(list(mrelease_year.keys())[0],max_count))
```
