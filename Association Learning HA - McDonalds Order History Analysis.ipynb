```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_csv("mcd.csv", index_col=None, header=None)
df.head()
```

```python
df.columns
datanew=df
datanew.columns
datanew.columns=['Invoice  number','Purchased','Also buyed','Also buyed1','Also buyed2']
print(datanew.head())
a=datanew.values.tolist()
x=0
for item in a:
    item=item[1:]
    a[x]=item
    b=str(item[3])
    if b=='nan':
        item=item[:3]
        a[x]=item
        c=str(item[2])
        if c=='nan':
            item=item[:2]
            a[x]=item
    x+=1
    #break
print(a[:25])
```

```python
te=TransactionEncoder()
te=te.fit(a) 
te_ary=te.transform(a) 
print(te_ary)
print(te.columns_)
df=pd.DataFrame(te_ary, columns=te.columns_)

frquent_itemsets=apriori(df, min_support=0.1,use_colnames=True)
frquent_itemsets

rules=association_rules(frquent_itemsets,metric= 'support',min_threshold=0.1)
rules
```
